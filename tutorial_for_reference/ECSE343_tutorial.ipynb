# Import libraries


```python
import numpy as np
import pickle
```

# Extract dataset and measurement



```python
with open('dataset_tutorial.pkl', 'rb') as f:
    loaded_data = pickle.load(f)
X = loaded_data['data']
y = loaded_data['target']

measurement = np.loadtxt('measurements_tutorial.csv', delimiter = ',')
```

```python
print(f"Input shape:          {X.shape}")
print(f"Target shape:         {y.shape}")
print(f"Measurement shape:    {measurement.shape}")
```

# Preprocess the data


```python
X = X.reshape((X.shape[0], -1)) #reshape the input data into a long vector
measurement = measurement.reshape(1,-1)   #reshapes the measurement data into exactly one row
```

```python
print(f"Input shape:          {X.shape}")
print(f"Target shape:         {y.shape}")
print(f"Measurement shape:    {measurement.shape}")
```

# Split dataset into train set and test set

```python
from sklearn.model_selection import train_test_split #import the train_test_split
```

```python
rng = np.random.seed(42) # create a random seed. this permits reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=rng)
```

```python
print(f"Train set input shape:          {X_train.shape}")
print(f"Test set input shape:           {X_test.shape}")
print(f"Train set target shape:         {y_train.shape}")
print(f"Test set target shape:          {y_test.shape}")
```

# Create a linear regression model

```python
from sklearn.linear_model import LinearRegression #import the Scikit learn LR library
```

```python
lr = LinearRegression()  #instantiate the linear regression model
```

# Train the linear regression model

```python
lr.fit(X_train,y_train) #fit the model to the training data
```

# Test the linear regression mode

```python
from sklearn.metrics import mean_squared_error #import an error metric
```

```python
pred = lr.predict(X_test) #get the predictions
```

```python
print(f"Test set input shape:           {X_test.shape}")
print(f"Test set target shape:          {y_test.shape}")
print(f"Predictions shape:              {pred.shape}")
```

```python
print(f"MSE on test set: {mean_squared_error(pred, y_test)}") #get the mean-squared error of predictions vs true
```

# Validate the model

```python
R_pred = lr.predict(measurement)
```

```python
print(f"Resistance prediction: {(R_pred.item())} Ohms.")
```
